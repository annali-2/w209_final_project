```python
build a stacked bar chart using the attached dataset
alt.data_transformers.disable_max_rows()
chart = alt.Chart(species,title=alt.TitleParams("Stacked Bar Chart of Each Species in US National Parks",
                  subtitle=["Hover over a bar to see more information appear!"])
                  ).mark_bar().encode(
    x=alt.X('count():Q', scale=alt.Scale(zero=False), title='Number of Records (Count)'),
    y=alt.Y('Category:N', sort='-x', title='Species')
).interactive()

HTML(chart.to_html())

chart.mark_bar(tooltip=True).encode(alt.Color('Region', title='Region'))
```

```python
chart = alt.Chart(species_group).mark_bar().encode(
   column=alt.Column('Category'),
   x=alt.X('Park Name'),
   y=alt.Y('count'),
   color=alt.Color('Park Name')
).properties(
     width=500,
    height=300
 )

chart.display() 
```

```python
parks = pd.read_csv('data/national-parks.csv')
parks['park'] = parks['name'].apply(lambda s:' '.join(s.replace('National', '').replace('Park', '').strip().split(" ")))

parks_long = pd.read_csv('data/nationalparks.csv')
parks_long['details'] = parks_long['details'].apply(lambda s:s.replace('"', "").strip())
parks_long['park'] = parks_long['details'].apply(lambda s: ' '.join(s.split(" ")[2:]))
species['Observations'] = species['Observations'].fillna(0)
```

```python
max_cat = species.loc[species.groupby(['Park Name','Category'])['Observations'].idxmax().sort_values(ascending=False)].reset_index(drop=True)
```

```python
max_cat
```

```python
parks_full = pd.merge(parks, parks_long, on="park")
```

```python
final = pd.merge(max_cat, parks_full, how='left', left_on=['Park Name'], right_on=['name'])
final
```
